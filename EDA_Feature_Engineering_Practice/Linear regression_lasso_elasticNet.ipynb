```python
## Divide the datset into independent and dependent

X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.33, random_state=36)
```

```python
X_train.head()
```

```python
X.shape
```

```python
X_train.shape,X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
###
```

```python
scaler.fit(X_train)
```

```python
#scaler.transform(X_train)
```

```python
## here we use fit_transform to fit and change the value sof the parameter to apply the training to the model

X_train=scaler.fit_transform(X_train)
```

```python
X_train
```

```python
X_train.shape
```

```python
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(X_train,y_train) # model is have standardised data and the output data
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
## So in here Instead of computing MSE first we are supposed to do the prediction for the new data.

y_pred=regression.predict(X_test) 
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
import numpy as np

mse=mean_squared_error(y_test,y_pred)
print(mse)
mae=mean_absolute_error(y_test,y_pred)
print(mae)
print(np.sqrt(mae))
```

```python
## Accuracy r2 and adjusted r square

from sklearn.metrics import r2_score

```

```python
score=r2_score(y_test,y_pred)
```

```python
score
```

```python
#display adjusted R-squared

1 - (1-score)*(len(y)-1)/(len(y)-X.shape[1]-1)

```

```python
from sklearn.linear_model import Ridge

ridge=Ridge()
ridge.fit(X_train,y_train)
```

```python
y_pred=ridge.predict(X_test)

mse=mean_squared_error(y_test,y_pred)
print(mse)
mae=mean_absolute_error(y_test,y_pred)
print(mae)
print(np.sqrt(mae))
```

```python
from sklearn.linear_model import Lasso
```

```python
lasso=Lasso(alpha=20.0)
lasso.fit(X_train,y_train)
```

```python
y_pred=lasso.predict(X_test)
mse=mean_squared_error(y_test,y_pred)
print(mse)
mae=mean_absolute_error(y_test,y_pred)
print(mae)
print(np.sqrt(mse))
```

```python
from sklearn.linear_model import ElasticNet
```

```python
elasticnet=ElasticNet(alpha=20.0)
elasticnet.fit(X_train,y_train)
```

```python
y_pred=elasticnet.predict(X_test)
mse=mean_squared_error(y_test,y_pred)
print(mse)
mae=mean_absolute_error(y_test,y_pred)
print(mae)
print(np.sqrt(mse))
```
